```python
import io
import PIL.Image
import numpy as np
from keras_segmentation.models.fcn import fcn_32_resnet50
from keras_segmentation.predict import model_from_checkpoint_path
from keras_segmentation.predict import predict
from skimage import segmentation
import cv2
from skimage import color
import moviepy.editor as mvi
import time

# tick = time.perf_counter()
def get_fcn():
    model_seg = fcn_32_resnet50(
        n_classes=14,  input_height=480, input_width=832)
    model_seg.load_weights('/Users/sahilapte/Documents/project school stuff/acer/FastApi_yolov5_semanticSegmentation-main/semantic_weights/fcn')
    return model_seg


model_seg = get_fcn()

li = [(127, 127, 127),
      (140, 140, 210),
      (114, 114, 255),
      (156, 70, 231),
      (75, 183, 186),
      (0, 255, 170),
      (0, 85, 255),
      (0, 0, 255),
      (0, 255, 255),
      (184, 255, 169),
      (165, 160, 255),
      (128, 50, 0),
      (0, 74, 111),
      (255, 255, 255)]

classes = ["Black Background", "Abdominal Wall", "Liver", "Gastrointestinal Tract", "Fat", "Grasper", "Connective Tissue",
           "Blood", "Cystic Duct", "L-hook Electrocautery", "Gallbladder", "Hepatic Vein", "Liver Ligament", "noice"]

frameSize = (512, 512)

out = cv2.VideoWriter('output_video.avi',
                      cv2.VideoWriter_fourcc(*'DIVX'), 25, frameSize)


def function(fps, pathin):
    cap = cv2.VideoCapture(pathin)
    count = 0
    while (cap.isOpened()):
        # Capture frame-by-frame
        ret, frame = cap.read()
        if ret == True:
            #print('Read %d frame: ' % count, ret)
            if count % fps == 0:
                cv2.imwrite("tempimg.png", frame)
                input_image = PIL.Image.open("tempimg.png")
                output = input_image.rotate(180)
                output.save("tempimg.png")
                predict(
                    model_seg,
                    inp="tempimg.png",
                    out_fname="temp_seg.png",
                    overlay_img=False,
                    # show_legends=True,
                    class_names=classes,
                    colors=li
                )
                img = cv2.imread("tempimg.png")
                g = cv2.cvtColor(cv2.imread("temp_seg.png"), cv2.COLOR_BGR2GRAY)
                #g = PIL.Image.open("temp_seg.png").convert('L')
                res = segmentation.mark_boundaries(img, g, mode='thick', color=(
                    255, 255, 255), outline_color=(255, 255, 255))
                cv2.imwrite("tempimg.png", 255*res)
                img = cv2.imread("tempimg.png")
                img = cv2.resize(
                    img, frameSize, interpolation=cv2.INTER_NEAREST)
                out.write(img)
            count += 1
        else:
            break
    # When everything done, release the capture
    cap.release()
    cv2.destroyAllWindows()
    out.release()
    clip = mvi.VideoFileClip('output_video.avi')
    clip.write_videofile("result.mp4")


function(1,"./input.mp4")
# end = time.perf_counter()
# print(end-tick)
```
